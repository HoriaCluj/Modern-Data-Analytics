### **Bike-Traffic Prediction Model**

1. Concatenate 2020 to 2026 hourly datasets
2. Resolve problematic variables (datetime) and drop useless ones
3. Add Infrastructure Deficit & Population based on site_ID
4. Train-Test Split to keep 2026 as a training set
5. Train `XGBoost` Regression Model to predict hourly bike traffic 

```python
import os

# Centralized data directory path
DATA_DIR = "C:/Users/Horia/Desktop/KUL/Modern Data Analytics"
```

```python
import pandas as pd
import polars as pl
```

**Step 1:** Concatenating the 2020 to 2026 hourly datasets

```python
df_2020 = pd.read_csv(os.path.join(DATA_DIR, "2020_hourly_all.csv"))
df_2021 = pd.read_csv(os.path.join(DATA_DIR, "2021_hourly_all.csv"))
df_2022 = pd.read_csv(os.path.join(DATA_DIR, "2022_hourly_all.csv"))
df_2023 = pd.read_csv(os.path.join(DATA_DIR, "2023_hourly_all.csv"))
df_2024 = pd.read_csv(os.path.join(DATA_DIR, "2024_hourly_all.csv"))
df_2025 = pd.read_csv(os.path.join(DATA_DIR, "2025_hourly_all.csv"))
df_2026 = pd.read_csv(os.path.join(DATA_DIR, "2026_hourly_all.csv"))
```

```python
df_raw = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024, df_2025, df_2026], ignore_index=False)
```

**Step 2:** Resolve problematic variables, add new ones, drop useless ones

```python
df = df_raw.drop(['direction', 'install_date', 'visibility'], axis=1)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year

season_map = {12: 'Winter', 1: 'Winter', 2: 'Winter',
               3: 'Spring', 4: 'Spring', 5: 'Spring',
               6: 'Summer', 7: 'Summer', 8: 'Summer',
               9: 'Autumn', 10: 'Autumn', 11: 'Autumn'}
df['season'] = df['month'].map(season_map).astype('category')

school_holidays = [
    ('2020-04-06', '2020-04-19'), ('2020-07-01', '2020-08-31'), ('2020-10-26', '2020-11-01'), ('2020-12-21', '2021-01-03'),
    ('2021-02-15', '2021-02-21'), ('2021-04-05', '2021-04-18'), ('2021-07-01', '2021-08-31'), ('2021-11-01', '2021-11-07'), ('2021-12-27', '2022-01-09'),
    ('2022-02-28', '2022-03-06'), ('2022-04-04', '2022-04-17'), ('2022-07-01', '2022-08-31'), ('2022-10-31', '2022-11-06'), ('2022-12-26', '2023-01-08'),
    ('2023-02-27', '2023-03-05'), ('2023-04-03', '2023-04-16'), ('2023-07-01', '2023-08-31'), ('2023-10-30', '2023-11-05'), ('2023-12-25', '2024-01-07'),
    ('2024-02-12', '2024-02-18'), ('2024-04-01', '2024-04-14'), ('2024-07-01', '2024-08-31'), ('2024-10-28', '2024-11-03'), ('2024-12-23', '2025-01-05'),
    ('2025-03-03', '2025-03-09'), ('2025-04-07', '2025-04-20'), ('2025-07-01', '2025-08-31'), ('2025-10-27', '2025-11-02'), ('2025-12-22', '2026-01-04'),
    ('2026-02-16', '2026-02-22'), ('2026-04-06', '2026-04-19'), ('2026-07-01', '2026-08-31'),
]
df['is_school_holiday'] = 0
for start, end in school_holidays:
    mask = (df['datetime'] >= start) & (df['datetime'] <= end)
    df.loc[mask, 'is_school_holiday'] = 1

# 24H Lag Feature
df = df.sort_values(['counter_id', 'datetime'])
df['lag_24h'] = df.groupby('counter_id')['bike_count'].shift(24).fillna(0)

df = df.drop('datetime', axis=1)
```

Encoding the remaining categorical variables (to prepare them for XGBoost fitting):

```python
day_map = {'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3,
           'Friday': 4, 'Saturday': 5, 'Sunday': 6}
df['day_of_week'] = df['day_of_week'].map(day_map)

season_map = {'Winter': 0, 'Spring': 1, 'Summer': 2, 'Autumn': 3}
df['season'] = df['season'].map(season_map).astype(int)
```

**Step 3:** Adding the `infrastructure_deficit` and `population` variables by matching the `counter_id`

```python
sites = pd.read_csv(os.path.join(DATA_DIR, "complete_sites.csv"),
                    usecols=['counter_id', 'infrastructure_deficit', 'population'])

df = df.merge(sites, on='counter_id', how='left')
```

```python
df['is_belgian_public_holiday'] = df['is_belgian_public_holiday'].astype(int)
df['is_weekend'] = df['is_weekend'].astype(int)
```

Adding an interaction feature between hour and weekend, rush hours may disappear on weekends

```python
# Feature 1: counter_id is already in df, just needs to be added to the features list
# Feature 2: hour x is_weekend interaction
df['hour_x_weekend'] = df['hour'] * df['is_weekend']
```

```python
df.head()
```

**Step 4:** Train-Test Splitting, keeping 2026 as the test-set

```python
train_df = df[df['year'] != 2026]
train_df = train_df.drop('year', axis=1)
train_df.head()
```

```python
test_df = df[df['year'] == 2026]
test_df = test_df.drop('year', axis=1)
test_df.head()
```

**Step 5:** Gradient Boosting (**XGBoost** and **LightGBM**) to predict `bike_counts`, perhaps also **Random Forest**

```python
features = [
    'counter_id',                  
    'hour', 'day_of_week', 'month', 'season',
    'is_belgian_public_holiday', 'is_weekend',
    'is_school_holiday',           
    'hour_x_weekend',              
    'lag_24h',                     
    'latitude', 'longitude',
    'temperature_2m', 'apparent_temperature', 'wind_speed_10m', 'precipitation', 'snowfall',
    'infrastructure_deficit', 'population'
]

X_train = train_df[features]
y_train = train_df['bike_count']

X_test = test_df[features]
y_test = test_df['bike_count']
```

a) `XGBoost Poisson Model`

```python
from xgboost import XGBRegressor

xgb_reg = XGBRegressor(
    objective="count:poisson", 
    n_estimators=500, 
    learning_rate=0.05, 
    max_depth=6, 
    subsample = 0.8,
    colsample_bytree = 0.8,
    n_jobs = -1,
    random_state=1350)

xgb_reg.fit(X_train, y_train)
y_pred_xgb = xgb_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
mae  = mean_absolute_error(y_test, y_pred_xgb)
r2   = r2_score(y_test, y_pred_xgb)

print(f"RMSE : {rmse:.2f}")
print(f"MAE  : {mae:.2f}")
print(f"R²   : {r2:.4f}")
```

Second `XGBRegressor` Model attempt:

```python
xgb_bet = XGBRegressor(
    objective='count:poisson',
    n_estimators=800,       # more trees with lower lr
    learning_rate=0.03,     # lower = more accurate, needs more trees
    max_depth=8,            # try pushing deeper
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=5,     # prevents splits on very rare patterns
    reg_alpha=0.1,          # L1 regularization
    reg_lambda=1.0,         # L2 regularization (default is 1)
    random_state=1354,
    n_jobs=-1
)

xgb_bet.fit(X_train, y_train)
y_pred_bet = xgb_bet.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred_bet))
mae  = mean_absolute_error(y_test, y_pred_bet)
r2   = r2_score(y_test, y_pred_bet)

print(f"RMSE : {rmse:.2f}")
print(f"MAE  : {mae:.2f}")
print(f"R²   : {r2:.4f}")
```

b) `LightGBM Poisson Model`

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import accuracy_score

lgbm_model = LGBMRegressor(
    objective='poisson',
    n_estimators=500,
    learning_rate=0.05,
    num_leaves=63,        
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_samples=20, 
    random_state=1989,
    n_jobs=-1
)
lgbm_model.fit(X_train, y_train)
y_pred_gbm = lgbm_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred_gbm))
mae  = mean_absolute_error(y_test, y_pred_gbm)
r2   = r2_score(y_test, y_pred_gbm)

print(f"RMSE : {rmse:.2f}")
print(f"MAE  : {mae:.2f}")
print(f"R²   : {r2:.4f}")
```

Second `LigthGBM` attempt:

```python
lgbm_tuned = LGBMRegressor(
    objective='poisson',
    n_estimators=1000,       
    learning_rate=0.03,     
    num_leaves=127,          
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_samples=10,    
    reg_alpha=0.1,           
    reg_lambda=1.0,          
    min_split_gain=0.01,     
    random_state=1989,
    n_jobs=-1
)

lgbm_tuned.fit(X_train, y_train)
y_pred_lgbm_tuned = lgbm_tuned.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred_lgbm_tuned))
mae  = mean_absolute_error(y_test, y_pred_lgbm_tuned)
r2   = r2_score(y_test, y_pred_lgbm_tuned)

print(f"RMSE : {rmse:.2f}")
print(f"MAE  : {mae:.2f}")
print(f"R²   : {r2:.4f}")
```
